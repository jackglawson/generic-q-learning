```python
from game import Game, GameData
from strategy import Strategy, PureState, State
from settings import learning_params, settings
import matplotlib.pyplot as plt
import numpy as np
```

```python
strategy = Strategy(learning=True)
```

```python
learning_params.num_epochs = 300

def learn(strategy):
    for epoch in range(learning_params.num_epochs):
        game = Game(strategy)
        game.play()
        if epoch % 100 == 0:
            print(epoch)
            
    return strategy
```

```python
def apply(strategy):
    learning_params.random_action_rate = 0
    game = Game(strategy)
    game.play()
    
    return game
```

```python
strategy = learn(strategy)
```

```python
settings.narrate_game = True
game = apply(strategy)
```
